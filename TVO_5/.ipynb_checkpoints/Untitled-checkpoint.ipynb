```python
# Поиск потенциально похожих пользователей
def get_similar_users(user_id, ratings_df):
    user_ratings = ratings_df.loc[ratings_df['User-ID'] == user_id]
    similar_users = pd.DataFrame(columns=user_ratings.columns)
    for item in user_ratings['ISBN']:
        #print(len(user_ratings))
        similar_users = pd.concat([similar_users, ratings_df.loc[(ratings_df['ISBN'] == item) & (ratings_df['User-ID'] != user_id)]])
    return similar_users.drop_duplicates(subset=['User-ID'])
```

```python
def get_similarity(user1, user2):
    common_values = np.array(list(set(user1['ISBN']) & set(user2['ISBN'])))
    user1 = user1[user1['ISBN'].isin(common_values)].sort_values(by=['ISBN'], ascending=True)
    user2 = user2[user2['ISBN'].isin(common_values)].sort_values(by=['ISBN'], ascending=True)
    if len(common_values) == 0:
        #print(f"Exit in ", 0)
        return 0
    correlation = np.corrcoef(user1['Book-Rating'].values, user2['Book-Rating'].values)[1, 0]
    if not np.isnan(correlation):
        return correlation
    else:
        return 1 - abs(user1['Book-Rating'].mean() - user2['Book-Rating'].mean())/10
```

```python
def user_based_prediction(user_id, ratings_df, item_id, k=1000):
    user = ratings_df.loc[ratings_df['User-ID'] == user_id]
    similarities = get_similar_users(user_id, ratings_df).apply(
        lambda x: get_similarity(ratings_df.loc[ratings_df['User-ID'] == x['User-ID']], user), axis=1)
    similarities.sort_values(ascending=False, inplace=True)
    similarities = similarities.nlargest(k)
    k_closest = ratings_df.iloc[similarities.index]
    k_closest = k_closest.loc[k_closest['ISBN'] == item_id]
    
    if len(k_closest) == 0:
        print("Ни один из ближайших соседей не оценил книгу с данным id")
        return np.nan

    similarities = similarities[k_closest.index]
    print(similarities)
    k_closest.loc[:, 'Book-Rating'] = 10
    print(k_closest)
    #print(k_closest.loc[k_closest['ISBN'] == item_id]['Book-Rating'])
    prediction = (user['Book-Rating'].mean() + 
    sum(similarities * (k_closest['Book-Rating'] - k_closest['Book-Rating'].mean()))/
    sum(similarities))
    return prediction
user_based_prediction(276746, ratings_df, '0786014512')
```

# Матричная факторизация

```python

reader = Reader(rating_scale=(0, 10))
ratings_dataset = Dataset.load_from_df(ratings_df, reader=reader)
train, test = train_test_split(ratings_dataset, test_size=0.15)

for alg in [SVD(), NMF()]:
    print(alg)
    alg.fit(train)
    predictions = alg.test(test)
    print(accuracy.rmse(predictions))
```
